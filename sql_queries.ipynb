1. find 10 top musics which have the most profit and their profit

```python
data = sqlite3.connect('Chinook_Sqlite.sqlite')
cursor = data.cursor()

query = """SELECT Track.Name, SUM(InvoiceLine.Quantity * InvoiceLine.UnitPrice) AS Total_Revenue FROM InvoiceLine
        JOIN Track ON InvoiceLine.TrackId = Track.TrackId
        GROUP BY Track.TrackId
        ORDER BY Total_Revenue DESC
        LIMIT 10"""

cursor.execute(query)

rows = cursor.fetchall()

for row in rows:
    print(f"{row[0]}: {Fore.RED}{row[1]}{Style.RESET_ALL}")
 
```

2. The most popular genre, the best in terms of number of songs sold and total revenue

```python
query = """WITH revenue_per_genre AS (
        SELECT Track.GenreId, COUNT(*) as SongsSold, SUM(InvoiceLine.Quantity * InvoiceLine.UnitPrice) AS TotalRevenue
        FROM Track
        JOIN InvoiceLine ON Track.TrackId = InvoiceLine.TrackId
        GROUP BY Track.GenreId
    )
    SELECT Genre.GenreId, Genre.Name, SongsSold, TotalRevenue
    FROM revenue_per_genre
    JOIN Genre ON revenue_per_genre.GenreId = Genre.GenreId
    ORDER BY TotalRevenue DESC
    LIMIT 1"""

cursor.execute(query)

row = cursor.fetchone()

print(f"The Most popular genre: {Fore.BLUE}{row[1]}{Style.RESET_ALL}")
print(f"Total of TheSongs sold: {Fore.GREEN}{row[2]}{Style.RESET_ALL}")
print(f"Total revenue: {Fore.RED}{row[3]}{Style.RESET_ALL}")
```

8. Cumulative number of songs sold each year individually

```python
query = """
WITH YearlySales AS (
    SELECT
        strftime('%Y', InvoiceDate) AS Year,
        COUNT(InvoiceLine.TrackId) AS TracksSold
    FROM
        Invoice
    JOIN
        InvoiceLine ON Invoice.InvoiceId = InvoiceLine.InvoiceId
    GROUP BY
        Year
    ORDER BY
        Year
),
CumulativeSales AS (
    SELECT
        Year,
        TracksSold,
        SUM(TracksSold) OVER (ORDER BY Year ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) AS CumulativeTracksSold
    FROM
        YearlySales
)
SELECT
    Year,
    CumulativeTracksSold
FROM
    CumulativeSales
ORDER BY
    Year;
"""

cursor.execute(query)

rows = cursor.fetchall()

for row in rows:
    print(f"Year {row[0]} : {Fore.YELLOW}{row[1]}{Style.RESET_ALL}")

```

9. Users whose total purchase is higher than the average total purchase of all users

```python
query = """
WITH CustomerTotal AS (
    SELECT
        CustomerId,
        SUM(Total) AS TotalPurchase
    FROM
        Invoice
    GROUP BY
        CustomerId
),
AverageTotal AS (
    SELECT
        AVG(TotalPurchase) AS AvgTotal
    FROM
        CustomerTotal
)
SELECT
    c.CustomerId,
    c.FirstName,
    c.LastName,
    ct.TotalPurchase
FROM
    CustomerTotal ct
JOIN
    Customer c ON ct.CustomerId = c.CustomerId
WHERE
    ct.TotalPurchase > (SELECT AvgTotal FROM AverageTotal);
"""

cursor.execute(query)

rows = cursor.fetchall()

for row in rows:
    print(f" {row[0]} {row[1]} {row[2]}, {Fore.CYAN}{row[3]:.2f}{Style.RESET_ALL}")
```
